```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
fig, ax = plt.subplots()
df=pd.read_csv("/content/ddata.csv")
points, = ax.plot(df['value'].to_list())
plt.plot(df['value'].to_list())
#plt.axis('off')
plt.show()
ax.axis([-1, 1000, -1, 350000])
x, y = points.get_data()
xy_pixels = ax.transData.transform(np.vstack([x,y]).T)
xpix, ypix = xy_pixels.T
width, height = fig.canvas.get_width_height()
ypix = height - ypix
lst = []
xc=[]
yc=[]
#print('Coordinates of the points in pixel coordinates')
for xp, yp in zip(xpix, ypix):
    #print ('{x:0.2f}\t{y:0.2f}'.format(x=xp, y=yp))
    xc.append(xp)
    yc.append(yp)
    #df2=pd.DataFrame(lst.append({'x': xp, 'y':yp}))
final={'x': xc, 'y':yc}
df2=pd.DataFrame(final)
df2.to_csv('pixel_coordinates.csv',index=False)
#fig.savefig('test.png', dpi=fig.dpi)
```
