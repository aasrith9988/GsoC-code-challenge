**Plotting ECG using Sci-Py ** 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
from scipy.misc import electrocardiogram
file_path = os.path.join(os.path.dirname('/content/ecgdata.csv'))
'''with np.load('/content/ecgdata.csv') as file:
  ecg = file["ecg"].astype(int)  
ecg = (ecg - 1024) / 200.0
return ecg
ecg = electrocardiogram()'''
ecg=pd.read_csv('/content/ecgdata.csv')
#ecg = ecg["100"].astype(int)  
ecg = (ecg - 1024) / 200.0
fs = 360
time = np.arange(ecg.size) / fs
plt.plot(time, ecg)
plt.xlabel("time in s")
plt.ylabel("ECG in mV")
plt.xlim(0, 2.2)
plt.ylim(-1.5, 2.0)
plt.show()
plt.axis('off')
print()
print()
'''print(ecg)
print(time.size)'''

plt.plot(time, ecg)
plt.xlabel("time in s")
plt.ylabel("ECG in mV")
plt.xlim(0, 2.2)
plt.ylim(-1.5, 2.0)
plt.show()
```
